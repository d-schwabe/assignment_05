## Step 4: Properly type the Oakland 311 data
### Copy the original dataframe

Before I convert the 2 columns, I'm going to copy the original dataframe into a new dataframe using `df.copy()`. This way I can still have the original dataframe saved in case I run into any problems later. I'll also be able to compare the edited dataframe with the original dataframe to check for errors. I'll name the copy dataframe simply `oakland_311`.

````python
oakland_311_copy = oakland_311_original.copy()
````

### Converting the date opened and closed columns to `datetime`

Now that I have the copy, I can convert the two columns from `object` dtype to `datetime` Dtype. I'll convert **DATETIMEINIT** first so I can compare the two columns. The syntax for this conversion is:

```python
df['column_name'] = pd.to_datetime(df['column_name'])
```

````python
oakland_311_copy['DATETIMEINIT'] = pd.to_datetime(oakland_311_copy['DATETIMEINIT'])
````

As noted in class, `oakland_311['Date_Opened']` is called a pandas **series**. Now I can use df.head() and df.info() to compare the converted DATETIMEINIT column to the DATETIMECLOSED column that I have not yet converted:

````python
oakland_311_copy.head()
````

````python
oakland_311_copy.info()
````

As we can see above, the format of the DATETIMEINIT column has changed, as has its Dtype. It is now a datetime64 Dtype.

Now I can convert the DATETIMECLOSED row:

````python
oakland_311_copy['DATETIMECLOSED'] = pd.to_datetime(oakland_311_copy['DATETIMECLOSED'])
````

I'll also take a look again to make sure the change has occurred:

````python
oakland_311_copy.head()
````

We can notice here (and indeed may have noticed before in the original oakland 311 file) that some number of the cells under the DATETIMECLOSED say 'NaN', which indicates that there is no data available. A quick glance across the table shows that this is in the case of cases that have a status of OPEN. Because of this missing data, we will not be able to find the difference between the DATETIMEINIT and DATETIMECLOSED columns for every case, but only for closed cases that have data for both the DATETIMEINIT and DATETIMECLOSED columns.

````python
oakland_311_copy.info()
````

### Using .min() and .max() to now find the correct earliest and latest entries
Now I can find the earliest date and the latest date of both columns, by performing `series.min()` and `series.max()` on these series.

````python
oakland_311_copy['DATETIMEINIT'].min()
````

````python
oakland_311_copy['DATETIMEINIT'].max()
````

````python
oakland_311_copy['DATETIMECLOSED'].min()
````

````python
oakland_311_copy['DATETIMECLOSED'].max()
````

The min and max tell us that the year 2009 is not complete (as of course is also true of 2021 because it's still happening). If I analyze the data by year later, I might consider excluding these years.

### Getting the difference of the two date columns

Using Pandas, I can find the difference of two dates by subtracting one datetime column from another. I'll now create a new column called **CLOSETIME** that shows how long it took for a case to be closed:

````python
oakland_311_copy['CLOSETIME'] = oakland_311_copy['DATETIMECLOSED'] - oakland_311_copy['DATETIMEINIT']
````

The resulting column should not be a `datetime` dtype, but rather a `timedelta` dtype. As a reminder from class, the term "delta" is often used to mean "change." I'll now use df.head() again to make sure the new column has been added:

````python
oakland_311_copy.head(40)
````

I used df.head(40) above to see more rows of the data, as I wanted to make sure that the operation worked for the rows for which there is data available for both DATETIMEINIT and DATETIMECLOSED. We see that the new CLOSETIME column has been created and that the operation seems to have worked for rows with complete data, such as rows 7 and 27.
